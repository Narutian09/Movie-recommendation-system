```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings=pd.read_csv("ratings.csv")
movies=pd.read_csv("movies.csv")
ratings=pd.merge(movies,ratings).drop(["genres","timestamp"],axis=1)
ratings.head()
```

```python
user_r=ratings.pivot_table(index=["userId"],columns=["title"],values="rating")
user_r.head()
```

```python
user_r=user_r.dropna(thresh=10,axis=1).fillna(0)
user_r.head()
```

```python
i_sim=cosine_similarity(user_r.T)
# i_sim
# i_similar=pd.DataFrame(i_sim,index=ratings.columns,columns=ratings.columns)
# print(i_sim)
i_sim=user_r.corr(method="pearson")
# i_similar
```

```python
i_sim.head(1060)
```

```python
def get_movies(movie_name,rating):
    similar_score=i_sim[movie_name]*(rating-2.5)
    similar_score=similar_score.sort_values(ascending=False)
    return similar_score
movies=[]
movie=input()
rating=int(input())
movies.append((movie,rating))
similar_movies=pd.DataFrame()
for movie,rating in movies:
    similar_movies=similar_movies.append(get_movies(movie,rating),ignore_index=True)
# similar_movies.head()
similar_movies.sum().sort_values(ascending=False)
recommendations=pd.DataFrame(similar_movies.T)
recommendations.drop([0],axis=1)[1:16]
```
